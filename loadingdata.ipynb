```python
import pandas as pd

Weather = pd.read_csv("4135691.csv", index_col = "DATE")
Weather.loc['2000-01-01':'2000-04-04']
```

```python
Weather.apply(pd.isnull).sum()/Weather.shape[0]
```

```python

#use the document to check out the important weather details
imp_weather = Weather[["PRCP", "SNOW", "SNWD", "TMAX", "TMIN"]].copy()
#making it easier for u to understand by renaming :D
imp_weather.columns = ["Precipitation","Snow","Snow_Depth","Temp_Max","Temp_Min"]
imp_weather.head()
                       
```

```python
#we gotta do smthin about these null values cz they bad for AI prediction
#also snow is mostly empty anyway we need to check its count I DID CHECK ITS COUNT IT GOT PERMA DELETED i restarted the kernel too :sad_face
imp_weather["Snow"].value_counts()
imp_weather["Snow_Depth"].value_counts() #not even once lol
```

```python
#it didnt even snow ONCE!
del imp_weather["Snow"]
del imp_weather["Snow_Depth"]
imp_weather.head()
```

```python
imp_weather["Precipitation"].isnull().sum()
```

```python
#imp_weather.loc["2000-01-01":"2015-04-04","Precipitation"]
#I THINK i found a better way to visualize
imp_weather[pd.isnull(imp_weather["Precipitation"])]
```

```python
#this shows a variance, not too many not too less, lets check it out
imp_weather["Precipitation"].value_counts()
```

```python
#i see lots of 0 in precipitation, so we can replace the null values by 0, as its a way way safer bet
imp_weather["Precipitation"] = imp_weather["Precipitation"].fillna(0)
```

```python
imp_weather["Precipitation"].isnull().sum()
```

```python
#YAYYYYYYY now checking for tempmin and max
imp_weather[pd.isnull(imp_weather["Temp_Max"])]
```

```python
#They r mostly NaN together and we can say that the prev or closest value will be the best approximation
imp_weather = imp_weather.ffill() #now temp min max both Nan will be assinged closest value 
```

```python
imp_weather.apply(pd.isnull).sum()/imp_weather.shape[0]
```
